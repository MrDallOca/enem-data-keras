```python
import json

with open('important_labels.json', 'r') as s:
    schema = json.load(s)
schema
```

```python
import numpy as np
from sklearn import preprocessing
import numbers
import math

for k,v in schema.items():
    min_max_scaler = preprocessing.MinMaxScaler()
    if v != False:
        if type(v) is list:
            c[k] = c[k].apply((lambda x: 
                       ((lambda l,y: l.index(y) if y in l else -1)(v,x) + 1) / (len(v))))
        else:
            ns = [n for n in c[k] if isinstance(n, numbers.Number) and not math.isnan(n)];
            mean = np.mean(ns);
            mmax = np.max(ns)
            mmin = np.min(ns)
            c[k] = c[k].apply(lambda x: (x - mean) / (mmax - mmin) if isinstance(x, numbers.Number) and not math.isnan(x) else 0)
    else:
        c = c.drop(k,axis=1)

c = c.dropna()
c
```
